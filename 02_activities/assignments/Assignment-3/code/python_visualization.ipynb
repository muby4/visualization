```python
import pandas as pd

# Load the data
df = pd.read_csv("ob_report_2025.csv", parse_dates=["Date Outbreak Began", "Date Declared Over"])

# View basic structure
print("Column names:\n", df.columns.tolist())
print("\nFirst 5 rows:")
print(df.head())

# Check data types and missing values
print("\nData Info:")
print(df.info())

# Count unique values per column to help decide what to plot
print("\nUnique values in each column:")
print(df.nunique())
```

```python
import matplotlib.pyplot as plt

# Count outbreaks by setting
outbreak_counts = df['Outbreak Setting'].value_counts()

print(outbreak_counts)
```

```python
# Plotting
plt.figure(figsize=(8,5))
outbreak_counts.plot(kind='bar', color='skyblue')

plt.title('Number of Outbreaks by Institution Setting')
plt.xlabel('Outbreak Setting')
plt.ylabel('Number of Outbreaks')
plt.xticks(rotation=45)
plt.tight_layout()

# Save the plot
plt.savefig("/Users/mubasharamohammed/visualization/02_activities/assignments/Assignment-3/images/outbreaks_by_setting_python.png", dpi=300, bbox_inches='tight')
# Show the plot
plt.show()

```
