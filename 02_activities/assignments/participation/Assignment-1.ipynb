```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import scipy
import PIL
import requests

np.random.seed(613)
x =	np.arange(50)
y1 = np.random.randint(0, 100, 50)
y2 = np.random.randint(0, 100, 50)
```

```python
#added a frame around the legend
# and changed the font size, number of columns, and shadow

fig, ax = plt.subplots(figsize=(5, 3))
ax.plot(x,y1, label = "person 1", color='green')
ax.plot(x,y2, label = "person 2", color='red')
ax.legend(loc ="lower right",
          frameon = True,
          fontsize = 12,
          ncol = 2,
          shadow = True,)
```

```python
#added legend outside the plot
fig, ax = plt.subplots(figsize=(5, 3))
ax.plot(x,y1, label = "person 1", color='green')
ax.plot(x,y2, label = "person 2", color='red')
ax.legend(loc ="upper left",
          frameon = True,
          fontsize = 8,
          ncol = 1,
          shadow = True,
          bbox_to_anchor=(1, 1)) #added bbox_to_anchor to move the legend outside the plot
```

```python
#create a new random seed for reproducibility
np.random.seed(613)
x1	= np.arange(50)
y1	= np.random.randint(0, 75, 50)
x2 = np.array(["Luffy", "Zoro", "Nami", "Usopp", "Sanji"])
y2 = np.array([110, 180, 240, 99, 220])
```

```python
#figure with mupltiple subplots

fig, (ax1, ax2) = plt.subplots(ncols = 2,
                       nrows =	1,
																							figsize = (7, 3))

ax1.scatter(x1,y1)
ax2.bar(x2, y2)
```

```python
#two different subplots in the same figure
#adding title, font	size, and color
fig, (ax1, ax2) = plt.subplots(ncols = 2,
																							nrows =	1,
																							figsize = (7, 3))
ax1.scatter(x1,y1, label = "person 1", color='purple', marker='^') #added label, color, and marker to the scatter plot
ax1.set_title("Scatter Plot", fontsize=14, color='gray', fontweight='bold') #added title to the scatter plot
ax1.set_xlabel("X Axis", fontsize=8, color='black') #added X axis label to the scatter plot
ax1.set_ylabel("Y Axis", fontsize=9, color='pink', fontweight='bold') #added Y axis label to the scatter	plot
ax1.legend(loc ="upper right") #added legend to the scatter plot


ax2.bar(x2, y2, label = "person 2", color='orange', hatch='//') #added label, color, and hatch to the bar plot
ax2.set_title("Bar Plot", fontsize=14, color='blue', fontweight='bold') #added title to the bar plot
ax2.set_xlabel("X Axis", fontsize=10, color='green') #added X axis label to the bar plot
ax2.set_ylabel("Y Axis", fontsize=11, color='red') #added Y	axis label to the bar plot
ax2.legend(loc ="upper right") #added legend to the	bar plot
```

```python
fig, someaxes = plt.subplot_mosaic([["ax1", "ax2"],["ax3", "ax2"]], figsize=(7, 4), layout='constrained') #layout added to get the title inside the plot

someaxes['ax1'].scatter(x1, y1, label = "person 1", color='purple', marker='^') #added label, color, and marker to the scatter plot
someaxes['ax2'].bar(x2, y2, label = "person 2", color='orange', hatch='//') #added label, color, and hatch to the bar plot
someaxes['ax3'].plot(x, y1, label = "person 1",	color='green', linestyle='--') #added label, color, and linestyle to the line plot

#addting big titles
someaxes['ax1'].set_xlabel('A Big Label', fontsize = 18)
someaxes['ax2'].set_xlabel('Another Big Label', fontsize = 18)
someaxes['ax3'].set_xlabel('Label2: 2 Fast 2 Furious', fontsize = 18)
```

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import requests
import scipy
import PIL

#two different types of plots on the same axes
x =	np.array(["Luffy", "Zoro", "Nami", "Usopp", "Sanji"])
y1	= np.array([110, 180, 240, 99, 220])
y2	= np.array([120, 170, 230, 100,	210])
fig, ax	= plt.subplots(figsize=(7, 3))
ax.bar(x, y1, label = "person 1", color='indigo')
ax.plot(x, y2, label = "person 2", color='red')
```

```python
y2_sd = np.std(y2) #calculate standard deviation for y2 values
fig, ax	= plt.subplots(figsize=(7, 3))
ax.plot(x, y2, color='red')	

ax.errorbar(x, #our x values
            y2, #our y values
            yerr = y2_sd, #our y error values
            fmt = 'none')
```

```python
y2_sd = np.std(y2) #calculate standard deviation for y2 values
fig, ax	= plt.subplots(figsize=(7, 3))
ax.plot(x, y2, color='red')	

ax.errorbar(x, #our x values
            y2, #our y values
            yerr = y2_sd, #our y error values
            fmt = 'none',
            color = 'blue', #error bar color
            linewidth = 2, #error bar line width
            capthick = 5,#error bar cap size
            capsize = 5) #error bar cap size
            #errorevery = 2)	#error bar error values if you want	to skip some points
```

```python
import plotly.graph_objects as go

x1 = np.array(['Luffy', 'Zoro', 'Nami', 'Usopp', 'Sanji'])
y1 = np.array([100, 180, 240, 99, 220])
```

```python
graph = go.Figure()

graph.add_trace(go.Bar(x=x1, y=y1))

graph.update_layout(title = 'Pirate Scores',
																			xaxis_title = 'Pirates',
                   yaxis_title = 'Scores')

graph.show()
```

```python
graph = go.Figure()

graph.add_trace(go.Scatter(x=x1, y=y1,
                       mode = 'markers',
                       marker = dict(size = 15,
                                     color = 'hotpink',
                                     opacity = 1,
                                     line = dict(width = 5,
                                                 color = 'purple'))))

graph.update_layout(title = 'Pirate Scores',
																			xaxis_title = 'Pirates',
                   yaxis_title = 'Score',
                   width = 500,
                   height	= 500,)

graph.show()
```

```python
from wordcloud import WordCloud
import pandas as pd


df = pd.read_csv("https://raw.githubusercontent.com/prasertcbs/basic-dataset/master/movie_quotes.csv",
                 on_bad_lines = 'skip') # to skip bad lines in the CSV file
df
```

```python
text = " ".join(each for each in df.quote)

wordcloud = WordCloud(background_color = "white",
                      colormap = "plasma").generate(text) # color map
                      
                      
fig,	ax = plt.subplots(figsize=(7, 3))
ax.imshow(wordcloud, interpolation = "bilinear")
ax.axis("off") # to turn off the axis
```

```python
import matplotlib.pyplot as plt
from matplotlib_venn import venn2, venn2_circles, venn2_unweighted

A = set(['apple', 'banana', 'watermelon'])
B = set(['pumpkin', 'blueberry', 'apple', 'key lime'])

diagram	= venn2([A, B], 
                set_labels=('Fruits', 'Pies'),
                set_colors=('blue', 'red'),
                alpha = 0.5) #set transparency

diagram.get_label_by_id('10').set_text("\n".join(A-B)) #left circle label for A
diagram.get_label_by_id('11').set_text("\n".join(A&B)) #right circle label for A and B
diagram.get_label_by_id('01').set_text("\n".join(B-A)) #right	circle label for B
```
